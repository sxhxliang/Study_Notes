```python
import seaborn as sns
from scipy.stats import kendalltau
from scipy.stats import spearmanr
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.jointplot(x='size',y='total_bill',data=tips,stat_func = kendalltau)
```

```python
sns.jointplot(x='size',y='total_bill',data=tips,stat_func = spearmanr)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,stat_func = kendalltau,kind='hex',color = '#4CB391')
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,stat_func = kendalltau,kind='reside',color = '#4CB391')
```

```python
import numpy as np
```

```python
X = np.linspace(1,10,100)
```

```python
Y = X**3 + 5*X**2 + 4*X 
```

```python
import pandas
data = pandas.DataFrame(np.c_[X,Y],columns=['X','Y'])
```

```python
sns.lmplot(x = 'X',y ='Y',data = data,fit_reg = False)
```

```python
import matplotlib.pyplot as plt
sns.plot(X,Y)
```

```python
xx = [1,2,3,4]
yy = [4,9,16,25]
plt.plot(xx,yy)
```

```python
sns.distplot(data['Y'],rug = True,hist = False,vertical = True,color = '#ed0dd9') 
```
